```python
import numpy as np
import random
import matplotlib.pyplot as plt
def obrabotka(x_y):
    num_out_of_line = []
    num =[]
    real_x_y = ""
    for i in range(len(x_y)):
        if x_y[i] in "0123456789":
            real_x_y += x_y[i]
            try:
                if x_y[i+1] not in "0123456789":
                   num_out_of_line.append(real_x_y)
                   real_x_y = ""
            except:
                   num_out_of_line.append(real_x_y)
    real_x_y = list(map(int, num_out_of_line))
    return real_x_y    
```
