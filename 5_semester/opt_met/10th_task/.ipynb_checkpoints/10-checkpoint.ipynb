```python
from pulp import LpMaximize, LpProblem, LpStatus, lpSum, LpVariable

# Create the model
model = LpProblem(name="small-problem", sense=LpMaximize)

# Initialize x1,x2
x1 = LpVariable(name="x1", lowBound=0, cat="Integer")
x2 = LpVariable(name="x2", lowBound=0, cat="Integer")

#constraints
model += (1.1 * x1 + x2 >= 5, "red_constraint")
model += (x1 + 0.2 * x2 <= 15, "blue_constraint")
model += (x2 <= 3, "yellow_constraint")
model += (0.7 * x1 <= 2, "green_constraint")

# objective function
model += lpSum([x1, (-3) * x2])

# Solve the problem
status = model.solve()
for var in model.variables():
     print(f"{var.name}: {var.value()}")
print(f"objective: {model.objective.value()}")
```
