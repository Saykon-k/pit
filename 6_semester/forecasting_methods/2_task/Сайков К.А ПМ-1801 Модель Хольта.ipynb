```python
import pandas as pd
import copy
```

### 1 Задание 

```python
data = pd.read_csv('ausair.csv')
info_data = list(data['Passengers'])
info_data_real = copy.deepcopy(info_data)
```

```python
def exp_smoothing(data,alfa):
    data = list(data)
    data.reverse()
    return sum([alfa * (1-alfa)**i *data[i] for i in range(len(data))])
```

```python
def req_form(data,alfa,s_o):
    return (1-alfa)*data+alfa*s_o
```

```python
data_new = info_data
data_prom = []
s_o = info_data[0]
for i in range(len(info_data)):
    prom = req_form(info_data[i],0.59,s_o)
    data_prom.append(prom)
#     print(prom)
    s_o = prom
```

```python
new_years = []
for i in range(9):
    prom = req_form(info_data[-1],0.59,s_o)
    new_years.append(prom)
    s_o = prom
    info_data.append(prom)
data_prom.extend(new_years)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12,5, forward=True)
plt.plot(info_data_real,label='Начальные данные')
plt.plot(data_prom,label='Простое экспоненциальное сглаживание')
plt.legend()
plt.show()
```
