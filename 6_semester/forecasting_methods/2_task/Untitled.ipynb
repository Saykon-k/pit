```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def holt_method_MSE(data, alpha, beta):
    n = len(data)
    y, l, b = np.zeros(n), np.zeros(n), np.zeros(n)
    values = data.values.reshape(1, n)[0]
    y[0] = values[0]
    l[0] = values[0]
    b[0] = values[1] - values[0]
    for i in range(1, n):
        
        y[i] = l[i - 1] + b[i - 1]
        l[i] = alpha*values[i] + (1 - alpha)*(l[i - 1] + b[i - 1])
        b[i] = beta*(l[i] - l[i - 1]) + (1 - beta)*b[i - 1]
        print(y[i])
    mse = 1/n * sum((values[1:] - y[:-1])**2)
    print(mse)
    return y
```

```python
passengers = pd.read_csv('ausair.csv', index_col='Year', parse_dates=True)
```

```python
holt_method_MSE(passengers, 0.59, 0.43)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(12,5, forward=True)
# plt.plot(info_data_real, label='Начальные данные')
plt.plot(holt_method_MSE(passengers, 0.59, 0.43),label='Линейный тренд')
plt.legend()
plt.show()
```
