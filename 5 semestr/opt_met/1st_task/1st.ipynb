```python
import cvxpy as cp
import numpy as np
import math 
# Generate data.
x1 = cp.Variable()
x2 = cp.Variable()
x3 = cp.Variable()

#   
constraints = [2*x1+2*x2<=90,
               x1+2*x3<=60,
               4*x2+2*x3<=120,
               2*x2+4*x3<=130,
               x1+x3<=40,
               x1>=0,
               x2>=0,
               x3>=0]

#  
obj = cp.Maximize(3*x1+5*x2+8*x3)


prob = cp.Problem(obj, constraints)

prob.solve() 

print("status:", prob.status)
print("optimal val:", np.round(prob.value, 5))
print("optimal var:\n(floor){0:5}x1 = {1}\n{0:12}x2 = {2}\n{0:12}x3 = {3}".format('',math.floor(x1.value), math.floor(x2.value) , math.floor(x3.value)) )
print("optimal var:\n{0:12}x1 = {1}\n{0:12}x2 = {2}\n{0:12}x3 = {3}".format('',np.round(x1.value, 5), np.round(x2.value, 5) , np.round(x3.value, 5)) )
```
